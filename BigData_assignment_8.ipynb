```python
import pandas as pd
```

```python
def df_index_reform(df):
    for col in df.columns:
        if col[:2] == '01':
            df.rename(columns = {col:'Gold' + col[4:]}, inplace=True)
        if col[:2] == '02':
            df.rename(columns = {col:'Silver' + col[4:]}, inplace=True)
        if col[:2] == '03':
            df.rename(columns = {col:'Bronze' + col[4:]}, inplace=True)
        if col[:1] == '№':
            df.rename(columns = {col:'#' + col[1:]}, inplace=True)

    df_gold = df.drop(['Totals'])
    return df_gold
```

df_index_reform 함수는 olympics.csv 파일의 컬럼을보기 쉽게 바꿔주는 함수입니다. 
그리고 max()함수를 쓸때 Totals가 잡히기때문에 아랫줄에 Totals칼럼을 삭제해서 리턴합니다.

```python
df = pd.read_csv('olympics.csv', index_col=0, skiprows = 1)
df.head()
```

그냥 불러오면 컬럼명은 위와 같습니다.

```python
df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
df.head()
```

df_index_reform 함수를 사용하면 위와 같은 표로 인덱스가 바뀝니다.

```python
def most_gold():
    df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
    contry = df[df['Gold'] == df['Gold'].max()].index[0]
    return contry[:contry.find('(')]
```

most_gold() 함수에 위에 만들어 놓은 df_index_reform() 함수를 사용하여 olympics.csv파일의 컬럼명을 바꿔주고 
df['Gold']와 df['Gold'].max()의 값이 같은 df컬럼 하나를 찾고 그 컬럼의 인덱스 명을 리턴합니다.
괄호 전까지 슬라이싱하여 나라이름만 리턴합니다.

```python
df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
df[df['Gold'] == df['Gold'].max()]
```

위는 하계 금메달이 제일 많은 나라의 컬럼입니다.

```python
def biggest_diff():
    df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
    df_gold_gap = pd.DataFrame(abs(df['Gold']-df['Gold.1']), index=df.index)
    contry = df_gold_gap[df_gold_gap[0] == df_gold_gap.max()[0]].index[0]
    return contry[:contry.find('(')]
```

df_gold_gap에 데이터 프레임의 형식으로 index는 나라명, 0번째 값은 하계 금메달에서 동계 금메달을 뺀 절대값을 넣고 df_gold_gap의 0번째 값이 max인 값인 칼럼을 찾고 그 컬럼의 인덱스 명을 리턴합니다.
괄호 전까지 슬라이싱하여 나라이름만 리턴합니다.

```python
df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
df_gold_gap = pd.DataFrame(abs(df['Gold']-df['Gold.1']), index=df.index)
df_gold_gap
```

```python
df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
df['Points'] = df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']*1
df.head()[['Gold.2','Silver.2','Bronze.2','Points']]
```

df를 불러와 Points를 금메달, 은메달, 동메달에 점수를 곱해 넣어주면 위와 같아집니다.
Gold.2, Silver.2, Bronze.2는 동계와 하계의 메달들을 합친 갯수이므로 여기다가 점수를 곱해줍니다.

```python
def make_points():
    df = df_index_reform(pd.read_csv('olympics.csv', index_col=0, skiprows = 1))
    df_all_points = df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']*1
    
    df_points = pd.Series(df_all_points, index = df.index)
    return df_points
```
