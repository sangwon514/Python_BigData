```python
import pandas as pd
df = pd.DataFrame(['D','C-','C+','B-','B','B+','A-','A','A+'],
                 index = ['poor','ok','ok','good','good','good','excellent','excellent','excellent'])
df.rename(columns={0:'Grades'}, inplace = True)
df
```

```python
from pandas.api.types import CategoricalDtype

#Nominal Scale
grades = df['Grades'].astype(CategoricalDtype(categories = ['D','C-','C+','B-','B','B+','A-','A','A+']))
grades.head()
```

```python
#Ordered Scale
grades = df['Grades'].astype(CategoricalDtype(categories = ['D','C-','C+','B-','B','B+','A-','A','A+'], ordered = True))
grades
```

```python
grades > 'B'
```

```python
import numpy as np
x = np.array([1,2,3,4,5,6,7,8])
y = x
plt.figure()
plt.plot(x,y,'or')
```

```python
colors = ['green'] * (len(x)-1)
colors.append('red')

plt.figure()
plt.scatter(x,y,c=colors)
```

```python
plt.figure()
plt.scatter(x[:3],y[:3],c='red',label='Small')
plt.scatter(x[3:],y[3:],c='blue',label='Large')
plt.ylabel('Grades') #그릴때 한번에 그리자
plt.xlabel('Number of success') #실시간 수정가능
plt.title('Relationship between success and grade')
plt.legend()
```

```python
#Line Plots
```

```python
linear_data = x
linear_data
```

```python
exponential_data = linear_data**2
plt.figure()
plt.plot(x,linear_data,'-o',x,exponential_data,'-o') #plot을 하면 LinePlot이 기본 marker -, . x, ^등 조합해서 사용가능
```

```python
plt.figure()
plt.plot(x,linear_data,'-o',x,exponential_data,'-o') 
plt.plot(x[:3],[22,44,55],'--r')
plt.legend(['Baseline','Completion','Test'])
```

```python
plt.figure()
plt.plot(x,linear_data,'-o',x,exponential_data,'-o') 
plt.plot(x[:3],[22,44,55],'--r')
plt.legend(['Baseline','Completion','Test'])
plt.fill_between(x,linear_data,exponential_data,facecolor = 'red', alpha=0.2)
```
