```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridespec

plt.figure()

gspec = gridespec.GridSpec(3,3)

lower_left = plt.subplot(gspec[1:,:2])
top_histogram = plt.subplot(gspec[0,:2], sharex = lower_left)
side_histogram = plt.subplot(gspec[1:,2], sharey = lower_left)
```

```python
X = np.random.normal(loc = 0.0, scale = 1.0, size=10000)
Y = np.random.normal(loc = 0.0, scale = 1.0, size=10000)

lower_left.scatter(X,Y, color = 'r')
top_histogram.hist(X, bins=100, color = 'b', density=True)
s = side_histogram.hist(Y, bins=100, color = 'b', density=True, orientation='horizontal')
```
