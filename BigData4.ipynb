```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6,7,8])
plt.figure()
plt.plot(x,'-o')
```

```python
plt.figure()
plt.subplot(1,2,1) ## 행,렬,인덱스

linear_data = x
plt.plot(linear_data,'-o')
```

```python
exponential_data = linear_data**2

plt.subplot(1,2,2)
plt.plot(exponential_data, '-o')
```

```python
plt.subplot(1,2,1) ## 이 subplot으로 이동한다는 의미
plt.plot(exponential_data, '-o') ##warning 뜬거
```

```python
plt.figure()
ax1 = plt.subplot(1,2,1) ## subplot의 리턴값이 있음 1행2열 1번인덱스에 위치를 갖는 값
plt.plot(linear_data,'-o')

ax2 = plt.subplot(1,2,2,sharey=ax1) ## subplot의 주소값도 갖고 plt 위치로 이동되는? 실행도 됌
plt.plot(exponential_data, '-o')
```

```python
plt.figure()
ax1 = plt.subplot(1,2,1) ## subplot의 리턴값이 있음 1행2열 1번인덱스에 위치를 갖는 값
plt.plot(linear_data,'-o')

ax2 = plt.subplot(1,2,2,sharey=ax1) ## subplot의 주소값도 갖고 plt 위치로 이동되는? 실행도 됌
plt.plot(exponential_data, '-x') ##데이터가 큰축에 맞춰서 그림이 그려짐
```

```python
## Histograms (빈도를 나타내는 그래프)
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]
```

```python
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0,scale=1.0,size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0,scale=1.0,size=sample_size)
    axs[n].hist(sample,bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
plt.figure()
Y = np.random.normal(loc=0.0, scale=1.0, size=10000) #평균, 분산, 크기
X = np.random.random(size=10000)

plt.scatter(X,Y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3,3)

top_histogram = plt.subplot(gspec[0,1:])
side_histogram = plt.subplot(gspec[1:,0])
lower_right = plt.subplot(gspec[1:,1:])
```

```python
lower_right.scatter(X,Y)
top_histogram.hist(X,bins=100)
s = side_histogram.hist(Y,bins=100,orientation='horizontal')
```

```python
top_histogram.clear()
top_histogram.hist(X, bins=100, density=True)
side_histogram.clear()
side_histogram.hist(Y, bins=100, orientation='horizontal', density=True)
side_histogram.invert_xaxis()
```
