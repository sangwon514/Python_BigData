```python
matchurl = f'https://api.nexon.co.kr/kart/v1.0/users/{access_id}/matches?start_date=&end_date= &offset=&limit=&match_types='

response2 = requests.get(matchurl, headers = header)
matchList = response2.json()
matchList
```

```python
df = pd.DataFrame(matchList['matches'][0]['matches'])
df
```

```python
ranking = []
for rank in df['player']:
    ranking += rank['matchRank']
```
